5.Get rid of the missing value:
a.Drop columns which has missing value more than 50% of the total data
b.Fill the missing value in agent with the mean
c.Fill the missing value in country, children with the most common name


```python
#5a
data.dropna(thresh=0.5*len(data),axis=1, inplace=False)
```

```python
#5b
data.loc[:,"agent"].fillna(data.loc[:,"agent"].mean())
```

```python
#5c
data.loc[:,"country"].fillna(data.loc[:,"children"].mode()[0])
```

```python
#5c
data.loc[:,"children"].fillna(data.loc[:,"children"].mode()[0])
```

6. Make category based on the lead time: Short, Medium, and Long

```python
data["category"] = pd.cut(data["lead_time"],bins=3,labels=["Short","Medium","Long"])
```

7.Visualize the arrival_date_month to see in what month the most visitors come


```python
arrDateMonthList = data["arrival_date_month"].tolist()
```

```python
print(arrDateMonthList)
```

```python
mp.figure(figsize=(12,12))
mp.hist(arrDateMonthList, color="b", rwidth=0.8)
mp.title("Arrival Date Month",fontdict = {"fontsize":20})

mp.xlabel("Month", fontdict = {"fontsize":18})
mp.ylabel("Number of Visitors", fontdict = {"fontsize":18})

mp.grid(axis="y", ls="--", alpha=0.8)

mp.show()
```
